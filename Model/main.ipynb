#### <i> Import Library Files </i>

```python
import pandas as pd
import warnings

# Ignore all warnings
warnings.filterwarnings('ignore')
```

#### <i> Load The Dataset </i>

```python
df = pd.read_csv("infolimpioavanzadoTarget.csv")
df
```

#### <i> Data Preprocessing </i>

```python
# List of columns to keep
columns_to_keep = ['date', 'open', 'high', 'low', 'close', 'adjclose', 'volume', 'ticker','TARGET']

# Filter the dataframe
df = df[columns_to_keep]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


# Convert 'date' column to datetime format (optional but useful)
df['date'] = pd.to_datetime(df['date'])

# Encode the 'ticker' column as it is categorical
df = pd.get_dummies(df, columns=['ticker'], drop_first=True)

# Define features and target
X = df.drop(['date', 'TARGET'], axis=1)  # Exclude 'date' and 'TARGET' columns from features
y = df['TARGET']  # Target column

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

#### <i> Data Visulization And Analysis </i>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Target class distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='TARGET', data=df, palette='viridis')
plt.title("Target Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

#### <i> Models </i>
#### <i> Logistic Regression </i>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import cross_val_score

# Logistic Regression
log_model = LogisticRegression(random_state=42, max_iter=1000)
log_model.fit(X_train, y_train)
y_pred_log = log_model.predict(X_test)
accuracy_log = accuracy_score(y_test, y_pred_log)
print("Logistic Regression Accuracy:", accuracy_log)
print("Logistic Regression Classification Report:\n", classification_report(y_test, y_pred_log))
```

#### <i> Random Forest Classifier </i>

```python
# Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42, n_estimators=100)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("Random Forest Accuracy:", accuracy_rf)
print("Random Forest Classification Report:\n", classification_report(y_test, y_pred_rf))
```

#### <i> XGBoost Classifier </i>

```python
# XGBoost Classifier
xgb_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_test)
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
print("XGBoost Accuracy:", accuracy_xgb)
print("XGBoost Classification Report:\n", classification_report(y_test, y_pred_xgb))
```

#### <i> Comparison Of All Models </i>

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import f1_score

# Collect model names, accuracies, and F1 scores
model_names = ["Logistic Regression", "Random Forest", "XGBoost"]
accuracies = [
    accuracy_log,
    accuracy_rf,
    accuracy_xgb
]
f1_scores = [
    f1_score(y_test, y_pred_log, average='weighted'),
    f1_score(y_test, y_pred_rf, average='weighted'),
    f1_score(y_test, y_pred_xgb, average='weighted')
]

# Create a DataFrame for visualization
comparison_df = pd.DataFrame({
    "Model": model_names,
    "Accuracy": accuracies,
    "F1 Score": f1_scores
})

# Plot the comparison
plt.figure(figsize=(10, 6))
sns.barplot(x="Model", y="Accuracy", data=comparison_df, palette="viridis")
plt.title("Model Accuracy Comparison", fontsize=16)
plt.ylabel("Accuracy", fontsize=12)
plt.xlabel("Model", fontsize=12)
plt.ylim(0, 1)
plt.show()

# Plot F1 scores comparison
plt.figure(figsize=(10, 6))
sns.barplot(x="Model", y="F1 Score", data=comparison_df, palette="viridis")
plt.title("Model F1 Score Comparison", fontsize=16)
plt.ylabel("F1 Score", fontsize=12)
plt.xlabel("Model", fontsize=12)
plt.ylim(0, 1)
plt.show()
```
